# Create the binned fields

```julia
using Pkg
Pkg.activate("../")
Pkg.resolve()
include("../scripts/CopepodsNN.jl")
include("../scripts/param.jl")
using JupyterFormatter
enable_autoformat()
doplot = false
```

We work with the small and the large copepod groups:

```julia
datafile1 = joinpath(dataprocdir, "Small_copepods_DINCAE_$(regionname).nc")
datafile2 = joinpath(dataprocdir, "Large_copepods_DINCAE_$(regionname).nc")
```

## Files and directories

```julia
if isfile(datafile1) & isfile(datafile2)
    @info("Data files already created")
else
    @error("Data files have to be created with read_copepods_data.ipynb")
end
```

## Read observations in the region of interest

```julia
lon1, lat1, dates1, datavalues1, obsid1, varname1 =
    CopepodsNN.read_data_nc(datafile1, domaincompute);
lon2, lat2, dates2, datavalues2, obsid2, varname2 =
    CopepodsNN.read_data_nc(datafile2, domaincompute);
@info("Working on variables $varname1 and $varname2");
```

```julia
data_num_valid1, data_mean1 = CopepodsNN.compute_data_num(
    lon1,
    lat1,
    dates1,
    datavalues1,
    longrid,
    latgrid,
    fielddates_monthly,
);

data_num_valid2, data_mean2 = CopepodsNN.compute_data_num(
    lon2,
    lat2,
    dates2,
    datavalues2,
    longrid,
    latgrid,
    fielddates_monthly,
);
```

### Write data number and mean to file

```julia
CopepodsNN.write_netcdf_datanum_grid(
    joinpath(datadir, "Small_copepods_DINCAE_datanum_mean2.nc"),
    collect(longrid),
    collect(latgrid),
    fielddates_monthly,
    data_num_valid1,
    data_mean1,
)

CopepodsNN.write_netcdf_datanum_grid(
    joinpath(datadir, "Large_copepods_DINCAE_datanum_mean2.nc"),
    collect(longrid),
    collect(latgrid),
    fielddates_monthly,
    data_num_valid2,
    data_mean2,
)
```
